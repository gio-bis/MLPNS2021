```python
import numpy as  np
import pylab as plt
```

```python
def mygauss(x, mu, sig):
  return 1/np.sqrt(2*np.pi)/sig * (np.exp(-(x-mu)**2/2/sig**2))

x = np.arange(100) #container of 100 variables
mu = 50 #single number
sig = 10
mygauss(x, mu, sig) #array of floating point number
```

```python
plt.plot(x, mygauss(x, mu, sig))
```
