```python
import numpy as np
import pylab as pl
#!pip install astropy
import astropy
from astropy import units as u

%pylab inline
```

```python
def plotsquare(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
```

```python
def plotcircle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
plotsquare(1, ax=ax)
plotcircle(1, ax=ax)
```

```python
def isin(x, y, l):
  return np.sqrt(x**2 + y**2) < l 
```
