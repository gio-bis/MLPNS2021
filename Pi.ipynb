```python
import numpy as np
import pylab as pl
#!pip install astropy
import astropy
from astropy import units as u

%pylab inline
```

```python
def plotsquare(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
```

```python
def plotcircle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  
```

```python
fig, ax = pl.subplots(figsize=(5, 5))
plotsquare(1, ax=ax)
plotcircle(1, ax=ax)
```

```python
def isin(x, y, l):
  return np.sqrt(x**2 + y**2) < l 
```

```python
myList = [10, 50, 100, 500, 1000, 10000]
N = myList[0]
l = 1
inner = 0
np.random.seed(123)
fig, ax = pl.subplots(figsize=(5, 5))
plotsquare(1, ax=ax)
plotcircle(1, ax=ax)
pl.xlabel("x")
pl.ylabel("y")
for i in range(1,N+1):
    x = np.random.uniform(-1, 1) 
    y = np.random.uniform(-1, 1) 
    if np.sqrt(x**2 + y**2) < l:
       plot(x, y,'ro')
       inner = inner + 1
    else:
      plot(x, y, 'bo')
print ("La frazione di punti interni è: {:.2f}".format(inner))
pi = 4 * inner / N
print ("Il valore di pi è: {:.2f}".format(pi))


  
```
